```python
import random
import matplotlib.pyplot as plt



n_A = 10000
n_B = 11000
n_C = 0
n_D = 0

Apoints = []
Bpoints = []
Cpoints = []
Dpoints = []
Tpoints = []

p = 0.9
cutoff_time = 15000

print(n_A, n_B, n_C, n_D)

for t in range(cutoff_time):
    if random.random()< p and n_A >0 and n_B >0:
        # 'reaction' production

        n_A -= 1
        n_B -= 1
        n_C += 1
        n_D += 2

    if t % 300 == 0:
        Apoints.append(n_A)
        Bpoints.append(n_B)
        Cpoints.append(n_C)
        Dpoints.append(n_D)
        Tpoints.append(t)

print(n_A, n_B, n_C, n_D)

plotA = plt.scatter(Tpoints, Apoints)
plotB = plt.scatter(Tpoints, Bpoints)
plotC = plt.scatter(Tpoints, Cpoints)
plotD = plt.scatter(Tpoints, Dpoints)
plotT = plt.scatter(Tpoints, Tpoints)

display(plotA)
```
