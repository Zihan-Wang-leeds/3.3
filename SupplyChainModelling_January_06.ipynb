This section contains the basics for modelling how some **discrete** entities can interact with each other. This may be various individuals of different types: each individual may be called an **agent**: e.g. a customer, a molecule, an individual during a pandemic, a rabbit, a bot etc, and there may be different types of agents, e.g. rabbits and foxes, that somehow 'react' (e.g. rabbit gets eaten by fox). This coupling between different types (i.e. foxes eating rabbits) leads to coupled dynamics and to types of **patterns** in the data that come up again and again in the real world. It does not matter what the example is about specifically, as this is a **completely general** modelling technique. You can adjust the ingredients to model different situations that are relevant to you.  Think of an example that could be relevant in a data science **work context** when you've got the hang of how this works. Some of the examples I will be using will be COVID/virus related, as it is topical, and doing relevant examples along those lines during the pandemic has given my previous students relevant and timely portfolio/CV material that has opened up **significant career accelleration opportunities**  for them. 

Imagine an **assembly line** (or a chemical reaction) set up like this: you start with parts A and B (say 10,000 and 11,000 each). At each time step, there is a certain probability (e.g. 9/10) that they get assembled together, resulting in one piece of product C and 2 bits of rubbish D. The below code calculates the random dynamics till it reaches a **steady state** (i.e. no further change with time) and plots all 4 species over this **time series**. Try to adapt the code to different scenarios if you like. This example is very boring and linear - how could you bring in some non-linearity to make things more interesting? Pay attention to how the time series data is **visualised** here. 

```python
import random
import matplotlib.pyplot  as plt

n_A = 10000
n_B = 11000
n_C = 0
n_D = 0
prob = 0.90 # Probability 9/10
time = 15000
Apoints = []
Bpoints = []
Dpoints = []
Cpoints = []
Tpoints = []

for i in range(time):
    if random.random()<prob and n_A > 0 and n_B >0:
        #perform reactions
        n_A -= 1
        n_B -= 1
        n_C += 1
        n_D += 2
    if time % 100 == 0:
        Apoints.append(n_A)
        Bpoints.append(n_B)
        Cpoints.append(n_C)
        Dpoints.append(n_D)
        Tpoints.append(i)
```

```python
plotA = plt.scatter(Tpoints,Apoints)
plotB = plt.scatter(Tpoints,Bpoints, color = 'red')
plotC = plt.scatter(Tpoints,Cpoints, color = 'green')
plotD = plt.scatter(Tpoints,Dpoints, color = 'yellow')
display(plotA,plotB,plotC,plotD)
```

We will look at more interesting processes / additional complications of this setup soon. These 'reaction'-type models cover a huge variety of different systems from supply chains, to chemistry, to marketing to epidemiological models. You can add in more complications in order to give the scope for modelling the complexities of the real world.  E.g. there can be back reactions; intermediates; the probability can depend on how many agents there are (e.g. number of encounters between individuals) etc. This is the most basic type of model relating different entities via interactions, e.g. customers and marketing interventions; chemical molecules; rabbits and foxes; viruses and cells susceptible to infection; bots and security; or any types of agents that can interact and affect each other's behaviour or population. 

If you like you can start with 10,000 clients and 11,000 friends of your clients that you have marketed towards via some **targeted marketing intervention** (i.e. they are in a primed state that might make them more susceptible to your marketing campaigns). Then there is a certain probability that you get **new clients** out of the friends you have marketed towards, as well as a new friends pool out of the friends of the new clients, and so on. 

These are some basic concepts in modelling. A lot of contemporary modelling in data science is done via **agent-based models** https://en.wikipedia.org/wiki/Agent-based_model, https://towardsdatascience.com/tagged/agent-based-modeling; often strong coupling between different types of agents leads to interesting and chaotic behaviour https://en.wikipedia.org/wiki/Complex_system!
